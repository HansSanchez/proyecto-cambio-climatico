# Parte 2: Consolidación de links
### Web Scraping a "Semana" para investigación referente a los cambios climáticos en Colombia

### Paso 1: Importación de librerias

```python
# Importación de librerias
import pandas as pd

# Definición de la clase
class ImportClass:
    
    # Función inicial o constructor
    def __init__(self):
        # lectura de los archivos
            ## encoding='utf-8' es por los caracteres especiales que puede tener los datos
        self.df_new_1 = pd.read_csv(r'df_news_1.csv', encoding='utf-8') 
        self.df_new_2 = pd.read_csv(r'df_news_2.csv', encoding='utf-8')
```

```python
# 1.1 Instancia del objeto
IC = ImportClass()
```

```python
# 1.2 Impresión para lectura de datos del (Módulo 1)
IC.df_new_1
```

```python
# 1.3 Impresión para lectura de datos del (Módulo 2)
IC.df_new_2
```

### Paso 2: Consolidación de Dataframes

```python
# 2.1 Consolidación en un único dataframe de todas las noticias para su procesamiento
df_consolidated_no_clean = pd.concat([IC.df_new_1, IC.df_new_2])

# 2.2 Impresión del DataFrame únificado
df_consolidated_no_clean
```

### Paso 3: Limpieza del Dataframe consolidado

```python
# 3.1. Limpieza del DataFrame según la columna de links df_consolidated_no_clean.drop_duplicates(subset=["headers_link"])
# 3.2. Eliminación de columna que se creo en por error "Unnamed: 0" .drop('Unnamed: 0', axis=1)
# 3.3. Reseteo de index por la acción de combinación de DataFrames .reset_index().drop(['index'], axis=1)
df_consolidated_clean = df_consolidated_no_clean.drop_duplicates(subset=["headers_link"]).drop('Unnamed: 0', axis=1).reset_index().drop(['index'], axis=1)
```

```python
# Impresión del Dataframe limpio y consolidado
df_consolidated_clean
```

```python
# 3.4 Generación de archivo .csv
df_consolidated_clean.to_csv("df_consolidated_clean.csv")
```
