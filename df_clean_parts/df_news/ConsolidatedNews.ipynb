```python
# Importación de librerias
import pandas as pd

# Definición de la clase
class ImportClass:
    
    # Función inicial o constructor
    def __init__(self):
        # lectura de los archivos
            ## encoding='utf-8' es por los caracteres especiales que puede tener los datos
        self.df_new_1 = pd.read_csv(r'df_news_clean_1.csv', encoding='utf-8') 
        self.df_new_2 = pd.read_csv(r'df_news_clean_2.csv', encoding='utf-8')
        self.df_new_3 = pd.read_csv(r'df_news_clean_3.csv', encoding='utf-8')        
        self.df_new_4 = pd.read_csv(r'df_news_clean_4.csv', encoding='utf-8')        
        self.df_new_5 = pd.read_csv(r'df_news_clean_5.csv', encoding='utf-8')        
        self.df_new_6 = pd.read_csv(r'df_news_clean_6.csv', encoding='utf-8')        
        self.df_new_7 = pd.read_csv(r'df_news_clean_7.csv', encoding='utf-8')        
```

```python
# 1.1 Instancia del objeto
IC = ImportClass()
```

```python
# 2.1 Consolidación en un único dataframe de todas las noticias para su procesamiento
df_news_consolidated_no_clean = pd.concat([
    IC.df_new_1,
    IC.df_new_2,
    IC.df_new_3,
    IC.df_new_4,
    IC.df_new_5,
    IC.df_new_6,
    IC.df_new_7    
])

# 2.2 Impresión del DataFrame únificado
df_news_consolidated_no_clean
```

```python
# 3.1. Limpieza del DataFrame según la columna de links df_consolidated_no_clean.drop_duplicates(subset=["headers_link"])
# 3.2. Eliminación de columna que se creo en por error "Unnamed: 0" .drop('Unnamed: 0', axis=1)
# 3.3. Reseteo de index por la acción de combinación de DataFrames .reset_index().drop(['index'], axis=1)
df_news_consolidated_clean = df_news_consolidated_no_clean.drop_duplicates(subset=["titles"]).drop('Unnamed: 0', axis=1).reset_index().drop(['index'], axis=1)
```
